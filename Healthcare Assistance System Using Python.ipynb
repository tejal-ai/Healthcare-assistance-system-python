```python
import pandas as pd

# Blood report function


def clean_and_display_blood_data():
    print("\n Loading Blood Report Dataset...")
    
    df_blood = pd.read_csv('/Users/Tejal/OneDrive/Documents/maharashtra_100_dataset.csv')

    #  Cleaning the Blood Dataset
    df_blood.drop_duplicates(inplace=True)
    df_blood.dropna(inplace=True)
    df_blood["BLOOD_SAMPLE_ID"] = df_blood["BLOOD_SAMPLE_ID"].astype(str).str.strip()
    df_blood["PINCODE"] = df_blood["PINCODE"].astype(str).str.strip()

    print("\n Blood Dataset Cleaned Successfully!\n")

    print(" Available Blood Records:")
    print(df_blood.to_string(index=False))

    # Blood Sample ID search
    sample_id = input("\n Enter Blood Sample ID: ").strip()
    result = df_blood[df_blood["BLOOD_SAMPLE_ID"] == sample_id]

    if result.empty:
        print("\n No report found for this Blood Sample ID!")
    else:
        print("\n BLOOD REPORT FOUND:\n")
        print(result.to_string(index=False))



#  HOSPITAL SEARCH FUNCTION USING BUDGET RANGES


def clean_and_search_hospitals():
    print("\n Loading Hospital Dataset...")
    
    df_hos = pd.read_csv('/Users/Tejal/OneDrive/Documents/maharashtra_hospitals_updated python miniproject.csv')

    #  Cleaning Hospital Dataset
    df_hos.drop_duplicates(inplace=True)
    df_hos.dropna(inplace=True)
    df_hos["Registration Fee"] = pd.to_numeric(df_hos["Registration Fee"], errors="coerce")
    df_hos["Pin Code"] = df_hos["Pin Code"].astype(str).str.replace(".0","",regex=False).str.strip()

    print("\nHospital Dataset Cleaned Successfully!\n")

    print(" Available PIN Codes:")
    print(df_hos["Pin Code"].unique())

    user_pincode = input("\nEnter Your Nearby PIN CODE: ").strip()

    print("\n Select Your Budget Range:")
    print("1️⃣  ₹0 – ₹500")
    print("2️⃣  ₹500 – ₹1000")
    print("3️⃣  ₹1000 – ₹5000")
    print("4️⃣  Above ₹5000")

    choice = input("\nEnter choice (1/2/3/4): ")

 
    if choice == "1":
        filtered = df_hos[df_hos["Registration Fee"] <= 500]
    elif choice == "2":
        filtered = df_hos[(df_hos["Registration Fee"] > 500) & (df_hos["Registration Fee"] <= 1000)]
    elif choice == "3":
        filtered = df_hos[(df_hos["Registration Fee"] > 1000) & (df_hos["Registration Fee"] <= 5000)]
    elif choice == "4":
        filtered = df_hos[df_hos["Registration Fee"] > 5000]
    else:
        print("\n⚠ Invalid option selected!")
        return

    #  Filter Hospitals by PINCODE
    result = filtered[filtered["Pin Code"] == user_pincode]

    print("\n================================================")

    if result.empty:
        print(" No hospitals found for your criteria!")
        print("\nℹ Try these PIN Codes instead:")
        print(df_hos["Pin Code"].unique())
    else:
        print(f" Hospitals near {user_pincode} in your budget:\n")
        print(result.to_string(index=False))

    print("================================================")

    
#  MAIN MENU SYSTEM


while True:
    print("\n============= MAIN MENU =============")
    print("1️⃣  View Blood Report & Search Patient")
    print("2️⃣  Search Budget-Friendly Hospitals")
    print("3️⃣  Exit")
    print("=====================================")

    user_choice = input("Enter Your Choice: ")

    if user_choice == "1":
        clean_and_display_blood_data()
    elif user_choice == "2":
        clean_and_search_hospitals()
    elif user_choice == "3":
        print("\n✔ Thank You for using the Healthcare System! ")
        break
    else:
        print("Invalid Input! Please try again.")


```
